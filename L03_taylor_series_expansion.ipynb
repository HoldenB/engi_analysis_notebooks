```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import solve_ivp
from scipy.special import factorial
```

```python
x = np.linspace(-10, 10, 2000)
y = np.sin(x)

plt.plot(x, y, 'k', linewidth=2)
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.grid(True)

# First-Order Taylor Expansion

P = [1, 0] # x + 0
y_T1 = np.polyval(P, x)
plt.plot(x, y_T1, 'b--', linewidth=1.2)

# Third-Order Taylor Expansion

P = [-1 / factorial(3), 0, 1, 0] # -(1/3!)x^3 + x + 0
y_T3 = np.polyval(P, x)
plt.plot(x, y_T3, 'r--', linewidth=1.2)

# Fifth-Order Taylor Expansion

P = [1 / factorial(5), 0, -1 / factorial(3), 0, 1, 0] # (1/5!)x^5 -(1/3!)x^3 + x + 0
y_T5 = np.polyval(P, x)
plt.plot(x, y_T5, 'g--', linewidth=1.2)

# Seventh-Order Taylor Expansion

# -(1/7)x^7 + (1/5!)x^5 -(1/3!)x^3 + x + 0
P = [-1 / factorial(7), 0, 1 / factorial(5), 0, -1 / factorial(3), 0, 1, 0]
y_T7 = np.polyval(P, x)
plt.plot(x, y_T7, 'm--', linewidth=1.2)

# Ninth-Order Taylor Expansion

# (1/9)x^9 -(1/7)x^7 + (1/5!)x^5 -(1/3!)x^3 + x + 0
P = [1 / factorial(9), 0, -1 / factorial(7), 0, 1 / factorial(5), 0, -1 / factorial(3), 0, 1, 0]
y_T9 = np.polyval(P, x)
plt.plot(x, y_T9, 'c--', linewidth=1.2)

# Final results as an overlay

plt.legend(['sin(x)', '1st order', '3rd order', '5th order', '9th order'])
plt.show()
```
